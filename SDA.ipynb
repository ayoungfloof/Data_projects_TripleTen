[Now, as the data is clean, aggregate data per user per period in order to have just one record per user per period. It should ease the further analysis a lot.]

````python
# Calculate the number of calls made by each user per month. Save the result.

# Step 1: Extract the month from 'call_date' and add it as a new column
calls['month'] = calls['call_date'].dt.month

# Step 2: Calculate the number of calls made by each user per month
monthly_calls_per_user = calls.groupby(['user_id', 'month']).agg({'id': 'count'}).reset_index()

# Step 3: Pivot the table to have months as columns and user_id as rows
pivot_calls_per_user = monthly_calls_per_user.pivot_table(index='user_id', columns='month', values='id', fill_value=0)

# Step 4: Rename the columns for clarity
pivot_calls_per_user.columns = ['Month_' + str(col) for col in pivot_calls_per_user.columns]

# Step 5: Add a 'Total_Calls' column to calculate the total calls for each user
pivot_calls_per_user['Total_Calls'] = pivot_calls_per_user.sum(axis=1)

# Step 6: View the result
print(pivot_calls_per_user.sample(5))
````

[Calculate the monthly revenue from each user (subtract the free package limit from the total number of calls, text messages, and data; multiply the result by the calling plan value; add the monthly charge depending on the calling plan). N.B. This might not be as trivial as just a couple of lines given the plan conditions! So, it's okay to spend some time on it.]

````python
# Calculate the monthly revenue for each user
import pandas as pd
import numpy as np
import math

# Step 1: Ensure the 'month' column exists for each dataset
calls['month'] = calls['call_date'].dt.month
messages['month'] = messages['message_date'].dt.month
internet['month'] = internet['session_date'].dt.month

# Step 2: Group by user and month to get the monthly totals for calls, messages, and internet
monthly_minutes_per_user = calls.groupby(['user_id', 'month'])['duration'].sum().reset_index(name='duration')
monthly_messages_per_user = messages.groupby(['user_id', 'month'])['id'].count().reset_index(name='num_messages')

# For internet, sum the MB used first, then convert to GB
monthly_internet_per_user = internet.groupby(['user_id', 'month'])['mb_used'].sum().reset_index(name='total_mb_used')

# Correctly convert MB to GB before applying math.ceil
monthly_internet_per_user['total_gb_used'] = (monthly_internet_per_user['total_mb_used'] / 1024).apply(math.ceil)

# Step 3: Merge all user activities (calls, messages, internet) on 'user_id' and 'month'
merged_data = pd.merge(monthly_minutes_per_user, monthly_messages_per_user, on=['user_id', 'month'], how='outer')
merged_data = pd.merge(merged_data, monthly_internet_per_user[['user_id', 'month', 'total_gb_used']], on=['user_id', 'month'], how='outer')

# Step 4: Fill missing values with 0
merged_data.fillna(0, inplace=True)

# Step 5: Merge with user plans and plan details
merged_data_with_plans = pd.merge(merged_data, users[['user_id', 'plan']], on='user_id', how='left')
merged_data_with_plan_details = pd.merge(merged_data_with_plans, plans, left_on='plan', right_on='plan_name', how='left')

# Step 6: Calculate the monthly revenue directly using vectorized operations
# Adjust for 'mb_per_month_included' and convert to GB (MB / 1024)
merged_data_with_plan_details['monthly_revenue'] = (
    merged_data_with_plan_details['usd_monthly_pay']  # Base fee
    + np.maximum(0, merged_data_with_plan_details['duration'] - merged_data_with_plan_details['minutes_included']) * merged_data_with_plan_details['usd_per_minute']  # Extra minutes
    + np.maximum(0, merged_data_with_plan_details['num_messages'] - merged_data_with_plan_details['messages_included']) * merged_data_with_plan_details['usd_per_message']  # Extra messages
    + np.maximum(0, merged_data_with_plan_details['total_gb_used'] - (merged_data_with_plan_details['mb_per_month_included'] / 1024)) * merged_data_with_plan_details['usd_per_gb']  # Extra GB
)

# Step 7: Select the relevant columns and display the result
final_result = merged_data_with_plan_details[['user_id', 'month', 'plan', 'duration', 'num_messages', 'total_gb_used', 'monthly_revenue']]
final_result
````

Null Hypothesis: The average revenue from users of the Ultimate and Surf calling plans is equal.

Alternative Hypothesis: The average revenue from users of the Ultimate and Surf calling plans differs.

Using t-test  to compare the means of two independent samples: the revenue from the Ultimate plan and the revenue from the Surf plan. Welch's t-test is appropriate because the variance between the two groups may differ.

````python
# Test the hypotheses

from scipy import stats

# Separate the data by plan from 'merged_data_with_plan_details'
ultimate_revenue = merged_data_with_plan_details[merged_data_with_plan_details['plan'] == 'ultimate']['monthly_revenue']
surf_revenue = merged_data_with_plan_details[merged_data_with_plan_details['plan'] == 'surf']['monthly_revenue']

# Perform a two-sample t-test (Welch's t-test assuming unequal variances)
t_stat, p_value = stats.ttest_ind(ultimate_revenue, surf_revenue, equal_var=False)

# Display the results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision based on alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("We reject the null hypothesis. There is evidence to suggest a significant difference in average revenue between Ultimate and Surf plans.")
else:
    print("We fail to reject the null hypothesis. There is not enough evidence to suggest a significant difference in average revenue between Ultimate and Surf plans.")
````
