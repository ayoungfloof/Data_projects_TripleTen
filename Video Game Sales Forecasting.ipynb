# Video Game Sales Analysis and Insights

### Project Overview
This project aims to analyze video game sales data, focusing on identifying patterns that determine game success. By examining sales trends, user ratings, and regional preferences, we hope to provide insights into potential future market opportunities.

### Objectives
- Analyze sales trends across different gaming platforms and genres.
- Understand the impact of user and critic ratings on game sales.
- Identify regional differences in platform and genre popularity.
- Formulate and test hypotheses about user ratings and game features.

### Dataset Source
The dataset used in this analysis comes from open sources such as Kaggle, covering video game sales, ratings, and reviews from 1980 to 2016.

### Data Description
- **Name**: The name of the game
- **Platform**: The gaming console/platform (e.g., Xbox, PlayStation)
- **Year_of_Release**: The year the game was released
- **Genre**: The genre of the game (e.g., Action, Sports)
- **NA_sales**: Sales in North America (in million USD)
- **EU_sales**: Sales in Europe (in million USD)
- **JP_sales**: Sales in Japan (in million USD)
- **Other_sales**: Sales in other countries (in million USD)
- **Critic_Score**: Average critic score (maximum of 100)
- **User_Score**: Average user score (maximum of 10)
- **Rating**: ESRB rating (e.g., E for Everyone, T for Teen)

*Note: Data for 2016 may be incomplete.*

````python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
````

````python
# Step 1 - open data file and asses data

# Load the dataset
file_path = '/datasets/games.csv'  # replace this with your dataset path if needed
df = pd.read_csv(file_path)

# Display general information about the dataset
df.info()

# Display the first few rows of the dataset
df.head()

# Detect both 'TBD' and 'tbd' in the dataset (case-insensitive)
for column in df.columns:
    tbd_count = df[column].astype(str).str.contains('(?i)tbd', na=False).sum()  # '(?i)' makes the search case-insensitive
    if tbd_count > 0:
        print(f"Column '{column}' has {tbd_count} 'TBD/tbd' values.")
        
# Display the first few rows of the dataset
df.head()
````

````python
# Step 2 Bullet 1 - converting column names to lowercase

df.columns = df.columns.str.lower()

print(df.columns)
````

````python
# Step 2 Bullet 2 - check and convert data types were necessary

# Convert year_of_release to integer (if possible)
df['year_of_release'] = pd.to_numeric(df['year_of_release'], errors='coerce', downcast='integer')

# Convert user_score to numeric, automatically turning 'TBD/tbd' into NaN
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')

# Convert rating to a categorical variable
df['rating'] = df['rating'].astype('category')

# Check the data types again
df.info()
    
````

### Columns Changed and Why:

**year_of_release**

***Original type***: object

***New type***: integer

***Reason for change***: The year_of_release column represents the year when a game was released, which is naturally a numerical value. Converting it to integer allows us to perform year-based calculations and comparisons. Any non-numeric values are set to NaN using the errors='coerce' argument, and then handled separately.


**user_score**

***Original type***: object

***New type***: float

***Reason for change***: The user_score column represents ratings given by users, which range from 0 to 10. However, some entries, such as "TBD", caused this column to be treated as an object instead of numeric. Since we need the column to be numeric for calculations such as averages, correlations, and other statistical measures, we convert it to float.
During this conversion, non-numeric values like "TBD" are coerced to NaN, which helps in avoiding errors during analysis. The "TBD" values indicate that the user scores had not yet been determined at the time of data collection, and treating them as missing values ensures that they don't interfere with the calculations.


**rating**

***Original type***: object

***New type***: category

***Reason for change***: The rating column contains the ESRB rating (e.g., "E", "M", "T"), which categorizes games based on content suitability for different age groups. Since these are discrete categories rather than continuous numeric data, converting this column to the category type allows us to handle the data in a more appropriate way.

Creating stacked bar chart to visualize the distribution of different ratings for each genre to see if specific rating dominates each genre to see if it is reasonable to assume the rating and fill in the missing data in the ratings column

````python
# Step 2 Bullet 3 - handling missing values #2

# Step 1: Group by 'genre' and 'rating' to see the distribution of ratings for each genre
rating_genre_distribution = df.groupby(['genre', 'rating']).size().unstack(fill_value=0)

# Step 2: Calculate the percentage of each rating within each genre
rating_genre_percentage = rating_genre_distribution.div(rating_genre_distribution.sum(axis=1), axis=0) * 100

# Step 3: Plot the results as a stacked bar chart
plt.figure(figsize=(12, 8))
ax = rating_genre_percentage.plot(kind='bar', stacked=True, colormap='tab20', ax=plt.gca())

# Add horizontal grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add labels and title
plt.title('Distribution of Ratings for Each Genre')
plt.xlabel('Genre')
plt.ylabel('Percentage of Games (%)')
plt.legend(title='ESRB Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
````

***Genres with Limited Rating Diversity***:

Genres like Puzzle and Sports display very little diversity in their ratings, with the E (Everyone) rating being almost exclusive. This suggests a high consistency in the type of content delivered within these genres.


***Conclusion***:

The distribution of ESRB ratings across genres reveals patterns that are likely influenced by the target audience and the nature of the gameplay. Genres like Adventure, Sports, and Puzzle cater to a younger or broader audience with predominantly E (Everyone) ratings, while genres like Action, Shooter, and Role-Playing often have more mature content and are more frequently rated M (Mature) or T (Teen). ### Step 5.1  - Testing hypotheses

***Null Hypothesis***: The average user ratings for Xbox One and PC platforms are equal.

***Alternative Hypothesis***: The average user ratings for Xbox One and PC platforms are not equal.

***Explanation - Formulated the null and alternative hypotheses***:
The null hypothesis (H0) assumes that there is no significant difference between the average user ratings of Xbox One and PC. This means that any observed difference is due to random chance.
The alternative hypothesis (H1) assumes that there is a significant difference between the average user ratings of Xbox One and PC, implying that the observed difference is not due to random chance.

***Significance level chosen***:
Setting alpha to 0.05. This means that we are allowing a 5% probability of rejecting the null hypothesis when it is actually true. A common choice in hypothesis testing, this level balances the risk of Type I error (false positive) and ensures that we can still detect significant differences.

````python
# Step 5.1: Hypothesis Testing - Average user ratings of Xbox One vs PC

# Filter the data for Xbox One and PC platforms using relevant dataset (2012-2016)
xbox_one_ratings = relevant_df[relevant_df['platform'] == 'XOne']['user_score'].dropna()
pc_ratings = relevant_df[relevant_df['platform'] == 'PC']['user_score'].dropna()

# Set the significance level (alpha)
alpha = 0.05

# Formulate the hypotheses
# Null Hypothesis (H0): The average user ratings of Xbox One and PC are the same.
# Alternative Hypothesis (H1): The average user ratings of Xbox One and PC are different.

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(xbox_one_ratings, pc_ratings, equal_var=False)

# Print the results
print("Hypothesis Testing: Xbox One vs PC")
print(f"t-statistic: {t_stat:.2f}, p-value: {p_value:.4f}")

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis: The average user ratings of Xbox One and PC are different.")
else:
    print("Fail to reject the null hypothesis: The average user ratings of Xbox One and PC are the same.")
````

### Conclusion for Hypothesis Testing: Xbox One vs PC

The results of the hypothesis test indicate that the average user ratings of Xbox One and PC are not significantly different, with a t-statistic of 0.60 and a p-value of 0.5490. Since the p-value is greater than the chosen significance level (0.05), we fail to reject the null hypothesis. This means that we do not have sufficient evidence to conclude that there is a significant difference between the user ratings of games on these two platforms. Therefore, it appears that user preferences or experiences between Xbox One and PC were similar during the period from 2012 to 2016.

````python
# Step 5.2: Hypothesis Testing - Average user ratings for Action vs Sports genres (2012-2016)

# Filter the data for Action and Sports genres using the relevant dataset (2012-2016)
action_ratings = relevant_df[relevant_df['genre'] == 'Action']['user_score'].dropna()
sports_ratings = relevant_df[relevant_df['genre'] == 'Sports']['user_score'].dropna()

# Set the significance level (alpha)
alpha = 0.05

# Formulate the hypotheses
# Null Hypothesis (H0): The average user ratings for Action and Sports genres are the same.
# Alternative Hypothesis (H1): The average user ratings for Action and Sports genres are different.

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(action_ratings, sports_ratings, equal_var=False)

# Print the results
print("Hypothesis Testing: Action vs Sports Genres")
print(f"t-statistic: {t_stat:.2f}, p-value: {p_value:.4f}")

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis: We have evidence that the average user ratings for Action and Sports genres are different.")
else:
    print("Fail to reject the null hypothesis: We do not have sufficient evidence to conclude that the average user ratings for Action and Sports genres are different.")
````
