```python
# Step 2.1: Splitting the data into training and validation sets
# This code divides each dataset into a training set (75%) and a validation set (25%).
# The training set will be used to train the model, and the validation set will evaluate its performance.


# Define a function to split data
def split_data(data):
    # Separate features and target variable
    features = data.drop(['id', 'product'], axis=1)
    target = data['product']
    # Split the data
    features_train, features_valid, target_train, target_valid = train_test_split(
        features, target, test_size=0.25, random_state=12345
    )
    return features_train, features_valid, target_train, target_valid

# Split the datasets for each region
features_train_0, features_valid_0, target_train_0, target_valid_0 = split_data(geo_data_0)
features_train_1, features_valid_1, target_train_1, target_valid_1 = split_data(geo_data_1)
features_train_2, features_valid_2, target_train_2, target_valid_2 = split_data(geo_data_2)

# Display the sizes of the training and validation sets for verification
print(f"Region 0: Training size: {features_train_0.shape}, Validation size: {features_valid_0.shape}")
print(f"Region 1: Training size: {features_train_1.shape}, Validation size: {features_valid_1.shape}")
print(f"Region 2: Training size: {features_train_2.shape}, Validation size: {features_valid_2.shape}")
```

```python
# Step 2.2: Train the model and make predictions for the validation set
# This code trains a Linear Regression model on the training set for each region
# and makes predictions on the validation set.

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Define a function to train the model and make predictions
def train_and_predict(features_train, target_train, features_valid):
    # Initialize the Linear Regression model
    model = LinearRegression()
    # Train the model on the training set
    model.fit(features_train, target_train)
    # Make predictions on the validation set
    predictions = model.predict(features_valid)
    return model, predictions

# Train the model and make predictions for each region
model_0, predictions_valid_0 = train_and_predict(features_train_0, target_train_0, features_valid_0)
model_1, predictions_valid_1 = train_and_predict(features_train_1, target_train_1, features_valid_1)
model_2, predictions_valid_2 = train_and_predict(features_train_2, target_train_2, features_valid_2)
```

```python
# Step 4: Function to Calculate Profit for Selected Wells
def calculate_profit(predictions, target, n_wells, cost_per_well, revenue_per_barrel):
    """
    Calculate profit for selected wells in a region.

    Parameters:
    predictions (array): Predicted reserves for the region.
    target (array): Actual reserves for the region.
    n_wells (int): Number of wells to select.
    cost_per_well (float): Cost of developing one well.
    revenue_per_barrel (float): Revenue per thousand barrels.

    Returns:
    float: Total profit from the selected wells.
    """
    # Combine predictions and target values into a single DataFrame
    results = pd.DataFrame({'predictions': predictions, 'actual': target})
    
    # Sort wells by predicted reserves in descending order
    selected_wells = results.sort_values(by='predictions', ascending=False).head(n_wells)
    
    # Calculate total revenue from the actual reserves of selected wells
    total_revenue = selected_wells['actual'].sum() * revenue_per_barrel
    
    # Calculate total cost of developing the selected wells
    total_cost = n_wells * cost_per_well
    
    # Calculate profit
    profit = total_revenue - total_cost
    return profit

# Define constants
n_wells = 200
cost_per_well = 500000
revenue_per_barrel = 4500

# Calculate profit for each region
profit_region_0 = calculate_profit(predictions_0, target_valid_0, n_wells, cost_per_well, revenue_per_barrel)
profit_region_1 = calculate_profit(predictions_1, target_valid_1, n_wells, cost_per_well, revenue_per_barrel)
profit_region_2 = calculate_profit(predictions_2, target_valid_2, n_wells, cost_per_well, revenue_per_barrel)

# Display results
print("Profit Calculation:")
print(f"Region 0 Profit: ${profit_region_0:.2f}")
print(f"Region 1 Profit: ${profit_region_1:.2f}")
print(f"Region 2 Profit: ${profit_region_2:.2f}")
```

```python
# Define a function to calculate profit for a set of wells
def calculate_profit(predictions, actual, revenue_per_thousand, cost_per_well, budget, num_selected_wells=200):
    """Calculate the profit for the top wells."""
    # Select the top wells based on predicted reserves
    selected_indices = predictions.argsort()[-num_selected_wells:]  # Top wells by predictions
    selected_reserves = actual.iloc[selected_indices]  # Corresponding actual reserves
    total_revenue = selected_reserves.sum() * revenue_per_thousand  # Revenue calculation
    total_cost = num_selected_wells * cost_per_well  # Cost calculation
    profit = total_revenue - total_cost
    return profit

# Updated bootstrapping function
def bootstrap_profit(predictions, actual, revenue_per_thousand, cost_per_well, budget, num_selected_wells=200, num_samples=1000):
    """Perform bootstrapping to evaluate profit and risk."""
    profits = []
    for _ in range(num_samples):
        # Sample 500 wells randomly
        sampled_indices = np.random.choice(len(predictions), size=500, replace=False)
        sampled_predictions = predictions[sampled_indices]
        sampled_actual = actual.iloc[sampled_indices]
        
        # Calculate profit for the top 200 wells out of the sampled 500
        profit = calculate_profit(
            sampled_predictions, 
            sampled_actual, 
            revenue_per_thousand, 
            cost_per_well, 
            budget, 
            num_selected_wells
        )
        profits.append(profit)
    
    # Convert profits to a NumPy array for further analysis
    profits = np.array(profits)
    # Calculate risk of loss
    risk_of_loss = (profits < 0).mean() * 100  # Express as a percentage
    # Return the profit statistics
    return profits.mean(), np.percentile(profits, 2.5), np.percentile(profits, 97.5), risk_of_loss

# Constants
budget = 100_000_000  # Total budget in dollars
cost_per_well = 500_000  # Cost per well in dollars
revenue_per_thousand = 4_500  # Revenue per thousand barrels

# Perform bootstrapping for each region
print("Bootstrapping Profit Analysis:")
for region, (predictions, actual) in regions.items():
    mean_profit, lower_ci, upper_ci, risk = bootstrap_profit(
        predictions, actual, revenue_per_thousand, cost_per_well, budget
    )
    print(f"\n{region}:")
    print(f"  Average Profit: ${mean_profit:,.2f}")
    print(f"  95% Confidence Interval: (${lower_ci:,.2f}, ${upper_ci:,.2f})")
    print(f"  Risk of Loss: {risk:.2f}%")
```
