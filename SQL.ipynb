```python
import pandas as pd

# Load datasets into DataFrames
df_01 = pd.read_csv('/datasets/project_sql_result_01.csv')
df_04 = pd.read_csv('/datasets/project_sql_result_04.csv')

# Display the first few rows of each dataset to confirm successful import
print("Dataset 1: Taxi Company Rides (November 15-16, 2017)")
display(df_01.head())

print("Dataset 2: Neighborhood Drop-offs (November 2017)")
display(df_04.head())

# Check data types to ensure correctness
print("\nData Types for df_01:")
print(df_01.dtypes)

print("\nData Types for df_04:")
print(df_04.dtypes)
```

```python
# Ensure correct data types for df_01
df_01['trips_amount'] = df_01['trips_amount'].astype(int)

# Ensure correct data types for df_04
df_04['average_trips'] = df_04['average_trips'].astype(float)

# Verify the changes
print("Data Types for df_01 after correction:")
print(df_01.dtypes)

print("\nData Types for df_04 after correction:")
print(df_04.dtypes)
```

Data Type Correction Complete
The data types for both datasets are now confirmed to be correct:

df_01:
company_name: object (string).
trips_amount: int64 (integer).
df_04:
dropoff_location_name: object (string).
average_trips: float64 (float).

```python
# Sort the neighborhoods by 'average_trips' in descending order and select the top 10
top_10_neighborhoods = df_04.sort_values(by='average_trips', ascending=False).head(10)

# Display the result
print("Top 10 Neighborhoods by Average Drop-Offs:")
display(top_10_neighborhoods)
```

```python
#Graph 1: Taxi Companies and Number of Rides

# Import Matplotlib for visualization
import matplotlib.pyplot as plt

# Graph 1: Taxi Companies and Number of Rides
plt.figure(figsize=(12, 6))
plt.bar(df_01['company_name'], df_01['trips_amount'], color='skyblue')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.title('Number of Rides by Taxi Company (November 15-16, 2017)', fontsize=14)
plt.xlabel('Taxi Company', fontsize=12)
plt.ylabel('Number of Rides', fontsize=12)
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```

```python
#Graph 2: Top 10 Neighborhoods by Average Drop-Offs

# Graph 2: Top 10 Neighborhoods by Average Drop-Offs
plt.figure(figsize=(10, 6))
plt.bar(top_10_neighborhoods['dropoff_location_name'], top_10_neighborhoods['average_trips'], color='orange')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.title('Top 10 Neighborhoods by Average Drop-Offs (November 2017)', fontsize=14)
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Average Number of Drop-Offs', fontsize=12)
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```
